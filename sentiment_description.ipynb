```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize
```

```python
# Only required on the first run
nltk.download('vader_lexicon')
nltk.download('punkt', force=True)
nltk.download('punkt_tab')
```

```python
# Instantiate our analyzer
sid = SentimentIntensityAnalyzer()
```

```python
# Load VADER lexicon
vader_lexicon = SentimentIntensityAnalyzer().lexicon
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/My Drive/cleaned_womens_orgs_WI_FB.csv')
```

```python
df['account.pageDescription'] = df['account.pageDescription'].astype(str)
```

```python
df['account.pageDescription_score'] = df['account.pageDescription'].apply(
    lambda x: sid.polarity_scores(x).get('compound', 0))
```
